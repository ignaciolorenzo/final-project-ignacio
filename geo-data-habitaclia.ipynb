```python
import pandas as pd
import math
from bs4 import BeautifulSoup
import requests
import nums_from_string as nfs
import numpy as np
from re import search
from random import randint
from time import sleep
import openpyxl
import xlrd
import lxml
import re
from datetime import datetime, timedelta
```

```python
url = "https://www.habitaclia.com/comprar-vivienda-en-X/selinmueble.htm"
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')
raw_provinces = soup.find_all('li')    
provinces = []

for rp in raw_provinces:
        try:
            province = rp.find('a').get('href').split('https://www.habitaclia.com/comprar-vivienda-en-')[1].split('/')[0]
            if province not in ['x', 'X']: provinces.append(province)
                
        except:
            pass
```

```python
df = pd.DataFrame(columns = ['town','n_properties','province'])

for p in provinces:
    #sleep(randint(2,4))
    url = "https://www.habitaclia.com/comprar-vivienda-en-"+p+"/listapoblaciones.htm"
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')

    raw_content = soup.find_all('li')    
    towns = []
    town = ""
    for rc in raw_content:
        try:
            #print('1')
            town = rc.find('a').get('href').split('https://www.habitaclia.com/viviendas-')[1].split('.')[0]
            print(town)
            n_properties = int(rc.find('span').get_text().replace('.',''))
            #print('2')
            if town not in towns:
                    #print('3')
                    towns.append(town)
                    #print('4')
                    new_row = {'town': town, 'n_properties': n_properties, 'province': p}
                    #print('5')
                    #print(new_row)
                    df = df.append(new_row, ignore_index=True)
        except:
            pass
```
